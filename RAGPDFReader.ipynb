```python
##data ingestion
from langchain_community.document_loaders import PyPDFLoader
from google.colab import drive
drive.mount('/content/drive')

# After mounting, use PyPDFLoader with the full path
from langchain.document_loaders import PyPDFLoader
loader = PyPDFLoader("/content/drive/My Drive/aiml.pdf")
docs = loader.load()
docs
```

# New Section

```python
#splitting
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
chunk_documents=splitter.split_documents(docs)
chunk_documents
```

# New Section

```python
##embedding
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
vectorstore = FAISS.from_documents(
    documents=chunk_documents,
    embedding=HuggingFaceEmbeddings()
)


```

```python
!pip install ipykernel
!pip install transformers

from transformers import AutoTokenizer, AutoModelForCausalLM



model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain.chains import ConversationalRetrievalChain
pipe = pipeline(
    "text-generation",  # Type of task
    model=model,        # Our loaded model
    tokenizer=tokenizer,# Our tokenizer
    max_length=1024,     # Maximum response length
    temperature=0.7,    # Creativity level (0=boring, 1=creative)
    top_p=0.95,        # Diversity of responses
    repetition_penalty=1.15  # Prevents repeating words
)
llm=HuggingFacePipeline(pipeline=pipe)
```

```python
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,  # Our language model
    retriever=vectorstore.as_retriever(search_kwargs={"k": 2}),  # FAISS retriever
    return_source_documents=True,  # Return source pages
    verbose=True  # Print debugging info
)
```

```python
def chat_with_pdf():
    chat_history = []
    print(" Hello! I'm ready to answer questions about your PDF. Type 'quit' to exit.")

    while True:
        # Get user question
        question = input("\nYour question: ")

        if question.lower() == 'quit':
            print("Goodbye!")
            break

        # Generate response
        result = qa_chain({
            "question": question,
            "chat_history": chat_history
        })

        # Print response
        print("\n🤖 Answer:", result["answer"])

        # Optionally print sources
        print("\n📚 Sources:")
        for doc in result["source_documents"]:
            print("- Page:", doc.metadata.get('page', 'Unknown page'))

        # Update chat history
        chat_history.append((question, result["answer"]))

# Start the chat
chat_with_pdf()
```
